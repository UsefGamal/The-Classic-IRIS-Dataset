## Modeling - Manual modeling 
Using the 4 feauters to model the output
     `if petal length < 2.5` ==> then it is a setosa

````python
#Using 1 feature to predict manualy
def single_feature_pred(petal_length):
    """Predicts the iris species given petal length """
    if petal_length < 2.5:
        return 0 #Setosa 
    elif petal_length < 4.8:
        return 1 #Versicolor
    else:
        return 2 #Verginica
````

````python
x_train['petal length (cm)']
````

````python
Manual_y_pred = np.array([single_feature_pred(val) for val in x_train['petal length (cm)']])
````

````python
manual_acc = np.mean(Manual_y_pred == y_train)
print(f"Manual Model Accuarcy: {manual_acc * 100:.2f}%")
````

### Using Cross-validation to evaluate the model

````python
from sklearn.model_selection import cross_val_score,cross_val_predict
````

````python
model = LogisticRegression(max_iter=200)
````

### Misclassifing points

````python
y_pred =  cross_val_predict(model,x_train,y_train,cv=5)
````

````python
y_pred
````

````python
correct_predictions = y_pred == y_train

incorrect_predictions = ~correct_predictions

x_train[incorrect_predictions]
````

````python
y_train
````

````python
df_pred = x_train.copy()
df_pred['target_name'] = y_train.map({0: 'setosa', 1: 'versicolor', 2: "virginica"})
df_pred['correct_predictions'] = correct_predictions
df_pred['prediction'] = y_pred
df_pred['prediction_label'] = df_pred['prediction'].map({0: 'setosa', 1: 'versicolor', 2: "virginica"})
df_pred.head()
````

````python
def plot_incorrect_preds(df_pred, x_axis, y_axis):
    fig, axs = plt.subplots(2, 2, figsize=(10,10))
    axs = axs.flatten()
    sns.scatterplot(x = x_axis, y = y_axis, data=df_pred, hue='prediction_label', ax=axs[0])
    sns.scatterplot(x = x_axis, y = y_axis, data=df_pred, hue='target_name', ax=axs[1])
    sns.scatterplot(x = x_axis, y = y_axis, data=df_pred, hue='correct_predictions', ax=axs[2])
    axs[3].set_visible(False)
    plt.savefig("./Images/incorrect_preds"+str(sum(df_pred.count())))
    plt.show()
````

````python
plot_incorrect_preds(df_pred,'petal length (cm)', 'petal width (cm)')
````

## Model Tuning 
Determine the `hyperparameters`  of the model that maximize the model performance

````python
#x_test = df_test.drop(columns=["target", "target_name"]).values
# x_train
#y_test = df_test["target"].values
````

````python
error_rate = []
accuracies = []
reg_params = np.arange(1,10,0.1)
#Finding optiml C 
for i in reg_params :
    
    model = LogisticRegression(max_iter= 200 , C = i)
    model.fit(x_train,y_train)
    pred_i = model.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
    accuracies.append(cross_val_score(model,x_train, y_train, cv = 5 , scoring  = "accuracy"))
    print(f"Accuracy :{np.mean(accuracies)* 100:.2f}%  for C = {i}")
````

````python
model = LogisticRegression(max_iter=200 , C = 10)
````

````python
model.fit(x_train, y_train)
````

````python
y_pred = model.predict(x_test)
````

````python
test_set_correct = y_pred == y_test 
accuracies = np.mean(test_set_correct)
print(f"Accuracy :{accuracies* 100:.2f}%")
````

````python
df_pred_test = x_test.copy()
df_pred_test['target_name'] = y_test.map({0: 'setosa', 1: 'versicolor', 2: "virginica"})
df_pred_test['correct_predictions'] = test_set_correct
df_pred_test['prediction'] = y_pred
df_pred_test['prediction_label'] = df_pred_test['prediction'].map({0: 'setosa', 1: 'versicolor', 2: "virginica"})
df_pred_test.head()
````

````python
plot_incorrect_preds(df_pred_test,'petal length (cm)', 'petal width (cm)')
````
